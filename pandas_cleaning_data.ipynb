```python
################################## Solution 2: Replace empty cells ##################################
```

```python
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
df
```

_______________________

```python
################################## Cleaning wrong format ##################################
```

```python
# change column name
df = df.rename(columns={' Date': 'Date'})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.to_string())
```

```python
# drop NaT Not a Time
df.dropna(inplace=True)
print(df.to_string())
```

```python
# drop non-useful column
## axis=1 indicate that we want to drop a column rather than a row. 
df = df.drop('Unnamed: 0', axis=1)
print(df.to_string())
```

------------------------

```python
################################## Cleaning wrong data ##################################
```

```python
# replacing values
## sets up a loop that will iterate over all the row indices in the DataFrame.
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120

print(df.to_string())
```
